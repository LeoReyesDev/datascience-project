## Clustering Jer치rquico Aglomerativo

Lo primero que haremos en correr el agoritmo para diferentes n칰meros de clusters:

A la hora de decidir el n칰mero de clusters conveniente para los datos, es conveniente usar la -Elbow Rule-. Esta regla consiste en correr el algor칤tmo para varios valores de k, y graficar estos valores contra los errores.
Se usa en la pr치ctica un valor de k cercano al punto de 'cambio de la convexidad'. 

V칠ase:  https://en.wikipedia.org/wiki/Determining_the_number_of_clusters_in_a_data_set

En este caso, vamos a considerar el valor k=2, ya que adem치s de ser un buen valor serg칰n el criterio mencionado, nos va a servir para analizar como el tipo de cancer se ve relejado o no en los 2 clusters por crear.

```python
# Creamos el modelo con k = 2
# Cluster usando k- means
from sklearn.cluster import AgglomerativeClustering
HClustering = AgglomerativeClustering(n_clusters=2, affinity="cosine", linkage="complete")
HC = HClustering.fit_predict(datos_escalados)
```

```python
# Otras estadisticas
HClustering_2 = AgglomerativeClustering(n_clusters=2, affinity="euclidean", linkage="ward")
HC_2 = HClustering_2.fit_predict(datos_escalados)
HC_2 = [0 if x==1 else 1 for x in HC_2]
```

## Visualizaci칩n

Una de las dificultades al crear algor칤tmos de clustering es la visualizaci칩n de los datos de altas dimensiones.

En la clase sklearn.manifold, tenemos el m칠todo llamado t-distributed Stochastic Neighbor Embedding, identificado como TSNE. Esta es una t칠cnica probabil칤stica que mantiene las estructuras locales de los datos en el espacio de incrustaci칩n de dimensi칩n menor. Nos genera una 'proyecci칩n' de los datos en en plano 2-dimensional.
V칠ase:


http://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html


https://www.quora.com/How-does-t-SNE-work-in-simple-words

```python
# Se crea incrustaci칩n de los datos en dos dimensiones
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter= 4000, random_state = 2)
Y = tsne.fit_transform(datos_escalados)
```

```python
# Se grafica clusters modelados vs los reales seg칰n variable -diagnostic-

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.scatter(Y[:,0],Y[:,1],  c=HC, cmap = "jet", edgecolor = "None", alpha=0.35)
ax1.set_title('Clusters por clustering jer치rquico aglomerativo')

ax2.scatter(Y[:,0],Y[:,1],  c = datos['diagnosis'], cmap = "jet", edgecolor = "None", alpha=0.35)
ax2.set_title('Clusters Actuales')
```

```python
from sklearn.metrics import confusion_matrix,classification_report

# Resultado con coseno y linkage completo
print(confusion_matrix(datos['diagnosis'],HClustering.labels_))
print(classification_report(datos['diagnosis'],HClustering.labels_))
```

```python
# Resultado con distancia euclidiana y ward
print(confusion_matrix(datos['diagnosis'],HC_2))
print(classification_report(datos['diagnosis'],HC_2))
```

De esto, se puede decir que los clusters realmente capturan el estado del cancer de las personas, pues los indicadores son altos. Note que se cometi칩 m치s ruido al incluir pacientes con cancer maligno en benigno que al contrario.

```python
from scipy.cluster.hierarchy import dendrogram, linkage

HC_fit = HClustering.fit(datos_escalados)

def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(HC_fit, labels=HC_fit.labels_)
plt.show()
```
